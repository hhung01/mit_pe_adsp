```python
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
import folium

print('Folium installed and imported!')
```

```python
def cleanup(inp):
  if inp[1] != "+":
    return ""
  new_char = ""
  for char in inp:
    if char in ["\xa0"]:
      new_char += " "
    elif char not in ["\u202a", "\u202c"]:
      new_char += char
  return new_char
```

```python
def getCode(phone):
  if "(" in phone:
    return phone[phone.index("(") + 1:phone.index(")")]
  return phone[:phone.index(" ")]
```

```python
filename='chat.txt'
phones = set()
with open(filename, 'r') as file:
  while line := file.readline():
    extract = line.split(":", 4)
    if len(extract) >= 3:
      phone = extract[2].split("]")
      if len(phone) > 1:
        #phones.add(phone[1].encode('ascii', 'ignore').strip())
        #phones.add(cleanup(phone[1].strip()))
        #phones.add(phone[1].strip())
        stripped = phone[1].strip()
        cleaned = cleanup(stripped)
        #print("'{}'".format(stripped), "==> '{}'".format(cleaned), "==>", len(cleaned))
        if len(cleaned) > 0:
          area_code = getCode(cleaned)
          #print("'{}' ==> '{}'".format(cleaned, area_code))
          phones.add(area_code)

len(phones)
```

```python
df = pd.DataFrame(list(phones))
df.columns = ['Code']
df.sort_values('Code', inplace=True)
df.reset_index(inplace=True, drop=True)
df.head()
```
