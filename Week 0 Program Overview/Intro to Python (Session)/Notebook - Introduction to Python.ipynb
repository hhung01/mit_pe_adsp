## <font color='Blue'> Dictionary

Dictionaries are incredibly useful data structures in Python. Think of them as real-life dictionaries: you have a word (the "key") and a definition (the "value").

In Python, dictionaries are a collection of key-value pairs. They're unordered, changeable, and do not allow duplicates - which means that each key must be unique. You can use any immutable data type as a key (like numbers or strings), and practically any data type as a value.

Here's an example of a dictionary:

```python
person = {
    "name": "Alice",
    "age": 25,
    "city": "New York"
}
```

In this dictionary, "name", "age", and "city" are the keys, and "Alice", 25, and "New York" are the corresponding values.

You can access the values in a dictionary by referring to the name of the key like this:

```python
print(person["name"])  # Will output: Alice
```

If you want to change a value in the dictionary, you can do so by referencing the key:

```python
person["age"] = 26  # Alice just had a birthday!
```

If you want to add a new key-value pair, you can do so like this:

```python
person["job"] = "Engineer"  # Alice is an Engineer
```

You can also remove key-value pairs using the `del` keyword:

```python
del person["city"]  # Alice doesn't want to share her city anymore
```

Dictionaries can also hold complex types like lists, tuples, and other dictionaries. This makes them really powerful for structuring and organizing data in your Python programs.

## <font color='Blue'> "for" Loop

Think of a `for` loop as a tool that lets you repeat a task over and over again, automatically. It's kind of like saying, "for each item in this group, do this task."

Let's consider an example of a shopping list.

Imagine you have a shopping list with some items you need to buy:

```python
shopping_list = ['milk', 'bread', 'eggs', 'butter']
```

You can use a `for` loop to go through each item in the list and print it out:

```python
for item in shopping_list:
    print(item)
```

This piece of code is essentially saying, "For each `item` in `shopping_list`, print the `item`."

When you run this code, Python looks at the first item in `shopping_list`, calls it `item`, and then does whatever you've told it to do in the body of the loop (in this case, `print(item)`). Python repeats this process for every item in the list. The output would be:

```
milk
bread
eggs
butter
```

Each loop through the list is called an iteration, and in every iteration, the variable `item` changes to the next item in the list.

You can also use `for` loops with a sequence of numbers using the `range()` function. For example, if you want to print the numbers 0 through 4, you can write:

```python
for i in range(5):
    print(i)
```

Here, `i` is a common variable name to use in such loops, and `range(5)` generates a sequence of numbers from 0 up to, but not including, 5. So, this loop will print out:

```
0
1
2
3
4
```

So, to put it simply, a `for` loop in Python lets you perform an action for each item in a list, or each number in a sequence. It's a super useful tool in programming that helps you automate repetitive tasks!

<font color='blue'>*Now in case we need to check for each type of car, we need to write the above code which will become very lengthy in case we have a huge amount of data. Hence, a loop can be used in that situation.*

````python
cars_ratings.keys()
````

````python
car_names= cars_ratings.keys()
# print all car names
for c in car_names:
  print(c)
````

**The *for* loop in Python is used to iterate over a sequence (list, tuple, string) or other iterable objects.
<br><br>Loop continues until we reach the last item in the sequence. The body of for loop is separated
from the rest of the code using indentation.**

````python
car_names
````

````python
# Get names of all the cars which are under performing
for car in car_names:
  if cars_ratings[car]<50:
    print(f"{car} is not performing well") # formatinf string f"{}"
````
